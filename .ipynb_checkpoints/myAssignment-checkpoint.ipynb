Import libraries and give them abbreviations

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

Open entire piano audio file as mono and store the original sampling frequency in a variable

```python
file = 'r1.wav'
sig , sr = librosa.load(file,mono=True,sr=None)
```

Calculate the length in seconds 

```python
length_s = sig.shape[0] / sr
print(length_s)
```

Plot signal

```python
plt.plot(sig)
```

Plot x2 with respect to time

```python
time=np.arange(0,sig.shape[0])/sr
```

```python
plt.plot(time,sig)
```

Scale x2 and replot with labels

```python
sig.shape
```

```python
max(abs(sig))
```

```python
excursion=max(abs(sig))
sigNew = sig / excursion
```

```python
max(abs(sigNew))
```

```python
plt.plot(time,sigNew)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('r0 Waveform')
```

```python
plt.plot(time,sigNew)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('r0 Waveform')
plt.xlim(0,1)
plt.ylim(-.18,.18)
```

Plot only a segment of signal (instead of just zooming in)

```python
plt.plot(time[1:44100],sigNew[1:44100])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('r0 Waveform')
```
